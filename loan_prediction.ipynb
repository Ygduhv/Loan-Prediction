```python
# Apply dummy encoding on purpose column
df=pd.get_dummies(data,columns=['purpose'],drop_first=True)  #data is name of dataframe
```

```python
X=df.drop('not.fully.paid', axis=1)
y=df['not.fully.paid']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
X_train
```

```python
X_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X_train, y_train)
```

```python
X_test
```

```python
y_pred=dtree.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
```

```python
print(tn,fp,fn,tp)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree1 = DecisionTreeClassifier(criterion='entropy')
dtree1.fit(X_train,y_train)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'max_depth':[2,3,4,5],
           'max_features':[1,3,5,10],
           'min_samples_split':[2,3,5,10],
           'criterion':['gini','entropy'],
           'min_samples_leaf':[1,2,3,4]
           }
grid=GridSearchCV(dtree, param_grid, verbose=2)
grid.fit(X_train,y_train)
```

```python
# Pass parameter as a dictionary 
param_grid={'max_depth':[2,3,4,5], #hyperparameters
           'max_features':[1,3,5,10],
           'min_samples_split':[2,3,5,10],
           'criterion':['gini','entropy'],
           'min_samples_leaf':[1,2,3,4]
           }
# Make the object of the class GridSearchCV
grid=GridSearchCV(dtree,param_grid,verbose=0)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
# Train decision tree 
from sklearn.tree import DecisionTreeClassifier
dtree21=DecisionTreeClassifier(criterion='entropy',
                               max_depth= 3,
                               max_features= 1,
                               min_samples_leaf=1,
                               min_samples_split= 3)
dtree21.fit(X_train, y_train)
```
